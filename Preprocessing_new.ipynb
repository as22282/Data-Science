```python
def process_csv_files(directory):
    import pandas as pd
    import os

    df_list = [] 

    # Loop through each file in the directory
    for filename in os.listdir(directory):
        # take all the file end with .csv
        if filename.endswith('.csv'):
            # Read the CSV file 
            df = pd.read_csv(os.path.join(directory, filename))
            
            # column which i can use
            df_new=df[['Gaze direction left X', 'Gaze direction left Y', 'Gaze direction left Z','Gaze direction right X', 'Gaze direction right Y', 'Gaze direction right Z','Pupil diameter left','Pupil diameter right','Eye position left X (DACSmm)','Eye position left Y (DACSmm)','Eye position left Z (DACSmm)','Eye position right X (DACSmm)','Eye position right Y (DACSmm)','Eye position right Z (DACSmm)','Gaze point left X (DACSmm)','Gaze point left Y (DACSmm)','Gaze point right X (DACSmm)','Gaze point right Y (DACSmm)','Recording timestamp', 'Gaze point X', 'Gaze point Y','Gaze event duration','Eye movement type index','Fixation point X', 'Fixation point Y']]
            df_new = df_new.replace(",", ".",regex=True)#change , with .
            #Convert categorial feature into float
            df_new[['Gaze direction left X', 'Gaze direction left Y', 'Gaze direction left Z','Gaze direction right X', 'Gaze direction right Y', 'Gaze direction right Z','Pupil diameter left','Pupil diameter right','Eye position left X (DACSmm)','Eye position left Y (DACSmm)','Eye position left Z (DACSmm)','Eye position right X (DACSmm)','Eye position right Y (DACSmm)','Eye position right Z (DACSmm)','Gaze point left X (DACSmm)','Gaze point left Y (DACSmm)','Gaze point right X (DACSmm)','Gaze point right Y (DACSmm)']] = df_new[['Gaze direction left X', 'Gaze direction left Y', 'Gaze direction left Z','Gaze direction right X', 'Gaze direction right Y', 'Gaze direction right Z','Pupil diameter left','Pupil diameter right','Eye position left X (DACSmm)','Eye position left Y (DACSmm)','Eye position left Z (DACSmm)','Eye position right X (DACSmm)','Eye position right Y (DACSmm)','Eye position right Z (DACSmm)','Gaze point left X (DACSmm)','Gaze point left Y (DACSmm)','Gaze point right X (DACSmm)','Gaze point right Y (DACSmm)']].astype(float)

            # Calculate standard deviation for each column
            df=df_new.std(axis=0)
            df_f = df.to_frame().transpose()#Convert column into row
            df_other = pd.read_csv(os.path.join(directory, filename), usecols=["Participant name"]).head(1)# add participant name
            df_concat = pd.concat([df_f, df_other], axis=1)#merged columns and participants name
            df_list.append(df_concat)

    # Concatenate all dataframes in the list into a single dataframe
    df=pd.concat(df_list)

    return df
```

```python
directory = 'EyeT/EyeT/'
df = process_csv_files(directory)#value pass on
```
